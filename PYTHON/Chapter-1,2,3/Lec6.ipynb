```python
# Heatmap()
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
```

```python
sns.heatmap(data)
plt.show()
```

```python
sns.heatmap(data,vmin=30,vmax=70)
plt.show()
```

```python
sns.heatmap(data,cmap="tab20b_r")
plt.show()
```

```python
sns.heatmap(data,annot=True)
plt.show()
```

```python
sns.heatmap(data,vmin=10,vmax=70,cmap="bwr",annot=True,linewidth=1,linecolor="black",cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```
