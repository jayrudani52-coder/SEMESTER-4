```python
import pandas as pd
```

```python
data={"calories":[420,380,390,400],
      "duration":[50,40,45,48]}
df=pd.DataFrame(data)
print(df)
print(df.loc[0])
print(df.loc[1])
print(df.loc[[0,1]])
print(df.iloc[0])

print(df["duration"])
print(df["calories"])

df=pd.DataFrame(data,index=["day1","day2","day3","day4"])
print(df)

print(df.loc["day1"])
# nahi chale  , print(df.iloc["day1"])
```

```python
data={"Name":["Alice","Bob","Charlie"],
      "Age":[25,30,35],
      "City":["New York","Los Angeles","Chicago"]}
df=pd.DataFrame(data)
print(df)
print(df.loc[0])
print(df.iloc[0])
print("*********")
print(df.loc[0:2])
print("*********")
print(df.iloc[0:2])

df.loc[(df.Age>20) & (df.City=="Chicago")]
print("****************************")
df.set_index("Name",drop=False,inplace=True)
print(df)
```

```python
data={"A":[1,2,3,4],"B":[5,6,7,8]}
df=pd.DataFrame(data,index=["r1","r2","r3","r4"])
df
df.loc["r2":"r4","B"]=99
print(df)
df.iloc[1:3,0]=0
print(df)
df["C"]=[10,20,30,40]
print(df)
df.loc[df["A"]==0,"new_col"]="zero"
print(df)
```
