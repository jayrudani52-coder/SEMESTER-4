```python
# sort_values
import pandas as pd
```

```python
df=pd.DataFrame({"Name":["Alice","Bob","Charlie","David"],
                 "Score":[88,92,86,90]})
df
```

```python
sorted_df=df.sort_values(by="Score")
sorted_df
```

```python
sorted_df=df.sort_values(by="Score",ascending=False)
sorted_df
```

```python
df.sort_index(ascending=False)
```

```python
df.set_index("Name",inplace=False)
```

```python
df.sort_index(ascending=False)
```

```python
# apply()
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return "pass" if x>=50 else "Fail"
result=s.apply(pass_fail)
result
```

```python
df=pd.DataFrame({"Physics":[60,70,80],
                 "Chemistry":[65,75,85],
                 "Maths":[70,80,90]})
result=df.apply(sum)
print(result)
```

```python
result=df.apply(sum,axis=1)
print(result)
```

```python
result=df.apply(lambda x:x.mean())
print(result)
```

```python
df["Total"]=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
def grade(Total):
    if Total>=250:
        return "A"
    elif Total>=200:
        return "B"
    else:
        return "C"
df["Grade"]=df["Total"].apply(grade)
df
```

```python
df=pd.DataFrame({"Width":[1,2],"Height":[3,4]})
def get_stats(row): 
    return [row["Width"]*row["Height"],2*(row["Width"]+row["Height"])]
df
```

```python
df.apply(get_stats,axis=1,result_type="expand")
```

```python
df.apply(get_stats,axis=1,result_type="reduce")
```

```python
df.apply(get_stats,axis=1,result_type="broadcast")
```

```python
def odd_values(x,y,z):
    return x+y+z
df=pd.DataFrame({"A":[1,2],"B":[10,20]})
df
```

```python
df1=df.apply(odd_values,args=(1,2))
df1
```

```python
def odd_values(x,m):
    return x*m
df=pd.DataFrame({"A":[1,2],"B":[10,20]})
df
```

```python
df1=df.apply(odd_values,m=10)
df1
```

```python
# replace()
df=pd.DataFrame({"Marks":[80,85,0,90],
                 "Attendance":[75,80,0,85],
                 "Grade":["A","B","F","A"]})
df
```
