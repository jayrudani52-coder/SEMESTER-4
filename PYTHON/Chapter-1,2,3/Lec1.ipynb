```python
import pandas as pd
```

```python
a=[1.7,7,2]
a1=(1.4,2,3,4,5)
myvar1=pd.Series(a1)
print(myvar1)
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
print(type(myvar))
print(myvar[0])
print(myvar["y"])
```

```python
s_name=["ram","shyam","radha","seeta","geeta"]
marks=[50,60,70,80,90]
myvar=pd.Series(marks,index=s_name,name="Student Result")
print(myvar)
```

```python
cal={"day1":400,"day2":380,"day3":430,"day4":370}
myvar=pd.Series(cal,index=["day1","day2"])
print(myvar)
```

```python
cal={"day1":400,"day2":380,"day3":430,"day4":400}
myvar=pd.Series(cal,name="Calories burn per day")
print(myvar)
print(myvar.size)
print(myvar.dtype)
print(myvar.name)
print(myvar.index)
print(myvar.values)
print(type(myvar.values))
print(myvar.is_unique)
```

```python
subs=pd.read_csv("subs.csv",squeeze=True)
print(subs)
print(type(subs))
print(subs.describe())
print(subs.min())
print(subs.max())
print(subs.median())
print(subs.sum())
print(subs.count())
```

```python
print(subs[subs>200])
print(subs[subs>200].size)
print(subs[subs==subs.max()].index)
```
